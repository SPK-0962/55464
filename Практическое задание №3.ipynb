```python
#1
'''
Загрузите данные из файла в объект DataFrame, Добавьте заголовки к
столбцам: «index», «year», «month», «day», «min_t», «average_t», «max_t»,
«rainfall».
'''

import pandas as pd
# данные по Рубцовску с 1936 по 2022 года
df = pd.read_table('wr147815a1.txt', sep = ';', 
                   names = ['index', 'year', 'month', 'day', 'min_t', 'average_t', 'max_t', 'rainfall'])
df
```

```python
#2
#Удалите столбец index.

df.drop(['index'], axis=1)
```

```python
#3
'''
Используя метод info(), ответьте на вопросы:
3.1. Есть ли в данных пропущенные значения?
3.2. В каком столбце данных больше всего пропущенных значений?

'''
import numpy as np
df.info()

df['min_t'] = df['min_t'].astype(str).str.replace(' ', '')
df['average_t'] = df['average_t'].astype(str).str.replace(' ', '')
df['max_t'] = df['max_t'].astype(str).str.replace(' ', '')
df['rainfall'] = df['rainfall'].astype(str).str.replace(' ', '')
df = df.apply(pd.to_numeric, errors='coerce')
print('\n')
df.info()
print('\n')
print(df.loc[:, df.columns[df.isna().any()]].isna().sum())
print('\n В столбцах есть пропуски. \n Больше всего в "min_t" минимальной температуре')
```

```python
#4
#В данных за какой год больше всего пропусков?

zeros_all = df.isnull().values.ravel().sum()
zeros_per_row = df.isnull().sum (axis= 1 )
probels = zeros_all.max()

print(f"Всего в данных {probels} пропусков")
print(f"Максимально пропущено в строке {zeros_per_row.max()} пропусков \nВ годах\n")

years = []

for index, row in df.iterrows():
    if (row.isnull().sum() == zeros_per_row.max()):
        years.append(row['year'])
#в одном году может быть нескольо месяцев, а в них несколь дней с пропусками
print(list(set(years)))
```

```python
#5
'''
Объедините столбцы «year», «month» и «day» в один столбец «date» в формате
гггг-мм-дд (2000-01-20). Данные в новом столбце должны иметь формат
datetime;
'''
import datetime

df['date']  = (pd.to_datetime(df['year'].astype(str) + '-' +
                                  df['month'].astype(str) + '-' +
                                  df['day'].astype(str)))
df.info()
```
